# Analyse adaptive CPG output

```python
# it's easiest to do pltting in jupyter...
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('data/adaptation_property--2021_05_14-12:36:19_PM.txt','rb') as f:
	spikes = pickle.load(f)
```

```python
plt.plot([spike.timestamp/1000 for spike in spikes], [spike.id for spike in spikes],'|')
plt.xlabel('time (ms)')
#plt.xlim([600, 700])
plt.ylabel('neuron id')
plt.show()
```

```python
num_neurons = 4
nids = np.arange(1, num_neurons+1)
duration = 2

pop_spikes = [spike for spike in spikes if spike.id in nids]
print(f'pop firing rate = {len(pop_spikes)/duration/len(nids)} Hz')
```

```python
num_neurons = 8
duration = 2
nids1 = np.arange(1, num_neurons+1)
nids2 = np.arange(num_neurons+1, 2*num_neurons+1)

pop1_spikes = [spike for spike in spikes if spike.id in nids1]
pop2_spikes = [spike for spike in spikes if spike.id in nids2]

print(f'pop 1 firing rate = {len(pop1_spikes)/duration/len(nids1)} Hz')
print(f'pop 2 firing rate = {len(pop2_spikes)/duration/len(nids2)} Hz')
```
